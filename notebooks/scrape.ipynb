```python
import os
import time
import requests
```

```python
# change working directory to the /Users/tildeidunsloth/Desktop/DatSci_25/DatSci_25
os.chdir('/Users/tildeidunsloth/Desktop/DatSci_25/DatSci25')
output_dir = "data/folketing_referater"
os.makedirs(output_dir, exist_ok=True)
```

```python
years = list(range(2000, 2025))
møder_per_år = 153 

for year in years:
    samling_id = f"{year}1"
    for møde in range(1, møder_per_år + 1):
        møde_id = f"m{møde}"
        url = f"https://www.folketingstidende.dk/samling/{samling_id}/salen/{møde_id}/{samling_id}_{møde_id}_referat.pdf"
        filename = os.path.join(output_dir, f"{samling_id}_{møde_id}.pdf")

        try:
            r = requests.get(url)
            if r.status_code == 200:
                with open(filename, 'wb') as f:
                    f.write(r.content)
                print(f"✔ Downloaded: {filename}")
            else:
                print(f"✖ Not found: {url}")
            time.sleep(1)  # rate limiting
        except Exception as e:
            print(f"Error with {url}: {e}")
```
